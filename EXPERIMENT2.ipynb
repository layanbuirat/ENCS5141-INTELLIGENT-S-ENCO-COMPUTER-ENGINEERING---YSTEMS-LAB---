```python
# لعرض رسمة واحدة
# matplotlib.pyplot: مكتبة تستخدم لإنشاء الرسوم البيانية.
#numpy: مكتبة تستخدم لإجراء العمليات الحسابية والعمل مع المصفوفات.
import matplotlib.pyplot as plt
import numpy as np
# generating a sinwwave signal
#t: مصفوفة تحتوي على قيم الزمن من 0 إلى 1 ثانية بفاصل 0.001 ثانية بين كل نقطة
t = np.arange(0, 1, 0.001)
#sig: مصفوفة تحتوي على قيم الإشارة الجيبية المقابلة لكل قيمة زمنية.
sig = np.sin(2 * np.pi * 2 * t)
#plt.plot(t, sig): يرسم الإشارة الجيبية (sig) مقابل الزمن (t).
plt.plot(t,sig)
# إضافة تسميات للمحاور
# LABLE OF X-AXIAS : TIME
plt.xlabel('Time (sec)')
#label of y- axias Amplitude
plt.ylabel('Amplitude')
#plt.grid(True): يعرض شبكة على الرسم البياني لتسهيل قراءة القيم.
plt.grid(True)
#plt.show(): يعرض النافذة التي تحتوي على الرسم البياني.
plt.show()
```

```python
# لعرض رسمتين فوق بعض
# generating two sine wave signals
t = np.arange(0, 2, 0.001)
sig1 = np.sin(2 * np.pi * 2 * t)
sig2 = np.sin(2 * np.pi * 2 * t - np.pi/6)
sig3 = np.sin(2 * np.pi * 2 * t + np.pi/4)
# إنشاء Subplots (رسمين بيانيين)
fig, axs = plt.subplots(2, 1)
# تحديد مواقع العلامات (Markers)
#marker_indices: مصفوفة تحتوي على مؤشرات (Indices) للنقاط التي سيتم وضع العلامات عليها. هنا، يتم وضع علامة كل 40 نقطة.
marker_indices = np.arange(0, len(t), 40)
#axs[0].plot(): يرسم الإشارة sig1 مقابل الزمن t مع وضع علامات دائرية (marker='o') كل 40 نقطة.
axs[0].plot(t, sig1, label='Signal 1', marker='o', markevery=marker_indices)
#يحدد حدود محور الزمن من 0 إلى 2 ثانية.
axs[0].set_xlim(0, 2)
# تسمية المحاور
axs[0].set_xlabel('Time (sec)')
axs[0].set_ylabel('Amplitude')
#يعرض شبكة على الرسم البياني
axs[0].grid(True)
#يعرض وسيلة إيضاح (Legend) للإشارة
axs[0].legend()
#يضيف عنوانًا للرسم البياني
axs[0].set_title('Sine Wave Signals')
#s ضع علامة مربع
axs[1].plot(t, sig2, label='Signal 2', marker='s', markevery=marker_indices)
# وضع علامة مثلث ^
axs[1].plot(t, sig3, label='Signal 3', marker='^', markevery=marker_indices)
#يحدد حدود محور الزمن من 0 إلى 2 ثانية.
axs[1].set_xlim(0, 2)
# تسمية المحاور
axs[1].set_xlabel('Time (sec)')
axs[1].set_ylabel('Amplitude')
axs[1].grid(True)
axs[1].legend()
#عنوان الرسمة الثانية
axs[1].set_title('Sine and Cosine Wave Signals')
#يضبط التباعد بين الرسوم البيانية لتجنب التداخل بين العناصر
fig.tight_layout()
#عرض الرسمات
plt.show()
```

```python
#2.1.2 Using Seaborn
#seaborn: مكتبة لإنشاء تصورات بيانية جميلة ومعقدة بسهولة.
#pandas: مكتبة للتعامل مع البيانات بشكل جدولي (DataFrames).
import seaborn as sns
import pandas as pd
# generating a sinwwave signal
#t: مصفوفة تحتوي على قيم الزمن من 0 إلى 1 ثانية بفاصل 0.001 ثانية بين كل نقطة.
t = np.arange(0, 1, 0.001)
#إنشاء الإشارات الجيبية (Sine و Cosine)
sin = np.sin(2 * np.pi * 2 * t)
cos = np.cos(2 * np.pi * 2 * t)
#df: إطار بيانات يحتوي على الأعمدة التاليةtime sin cos
#Creating a dataframe from the time, sin, and cos curves
df = pd.DataFrame({'time':t, 'sin':sin, 'cos':cos})
# Create a visualization
#sns.relplot(): دالة في seaborn لإنشاء تصورات علائقية (Relational Plots).
#data=df: يتم استخدام إطار البيانات df كمصدر للبيانات.
#kind="line": يحدد نوع الرسم البياني كرسم خطي (Line Plot).
sns.relplot(data=df,kind="line",x="time",y="sin").set(title='Sinwave')
```

```python
import pandas as pd

# قراءة البيانات
df = pd.read_csv("/content/ENCS5141Datasets/ENCS5141_Exp2_DescriptiveStatistics.csv")

# تنظيف البيانات (حذف القيم المفقودة)
df_cleaned = df.dropna()

# حساب النزعة المركزية
Mean = df_cleaned.mean()
Median = df_cleaned.median()
Mode = df_cleaned.mode().iloc[0]  # أول منوال فقط

# عرض النتائج كإطار بيانات
Central_Tendency = pd.DataFrame({
    'Mean': Mean,
    'Median': Median,
    'Mode': Mode
})

print("Central Tendency:\n", Central_Tendency)

# حفظ النتائج في ملف CSV
Central_Tendency.to_csv("Central_Tendency_Results.csv", index=False)
print("Results saved to 'Central_Tendency_Results.csv'")
```

```python
#Task 2.4: Explain how the mean, median, and mode measure central tendency by observing the distribution curves depicted in the graphs above.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# إنشاء عينة من البيانات
np.random.seed(42)  # لتثبيت العشوائية
data = np.random.normal(50, 10, 1000)  # توزيع طبيعي بمتوسط 50 وانحراف معياري 10

# تحويل البيانات إلى إطار بيانات (DataFrame)
df = pd.DataFrame(data, columns=["Values"])

# حساب النزعة المركزية
Mean = df["Values"].mean()
Median = df["Values"].median()
Mode = df["Values"].mode().values[0]  # أول منوال فقط

# إنشاء الرسم البياني
plt.figure(figsize=(10, 6))
sns.histplot(df["Values"], kde=True, color="blue", bins=30)

# إضافة خطوط للنزعة المركزية
plt.axvline(Mean, color='red', linestyle='--', label=f'Mean = {Mean:.2f}')
plt.axvline(Median, color='green', linestyle='--', label=f'Median = {Median:.2f}')
plt.axvline(Mode, color='purple', linestyle='--', label=f'Mode = {Mode:.2f}')

# إضافة عنوان وتسميات
plt.title("Distribution of Values with Central Tendency Measures", fontsize=16)
plt.xlabel("Values", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.legend()

# عرض الرسم البياني
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# قراءة البيانات (تأكد من المسار الصحيح)
df = pd.read_csv("/california_housing_train.csv")

# توحيد أسماء الأعمدة (أحرف صغيرة، بدون مسافات)
df.columns = df.columns.str.lower().str.strip()

# التحقق من أسماء الأعمدة
print("Columns in the dataset:", df.columns)

# الخطوة 1: التعامل مع القيم المفقودة في 'height' -----------------------------------

# ترميز الجنس (إذا كان فئويًا) لتحليل الارتباط
df['latitude'] = df['housing_median_age'].astype('category').cat.codes

# التحقق من الارتباط بين الارتفاع والجنس
correlation = df[['housing_median_age', 'latitude']].corr()
print("\nCorrelation between housing_median_age and latitude:\n", correlation)

# رسم توزيع الارتفاع لكل جنس
plt.figure(figsize=(8, 5))
sns.histplot(df, x="housing_median_age", hue="latitude", kde=True, bins=30)
plt.title("Housing Median Age Distribution by Latitude")
plt.xlabel("Housing Median Age")
plt.ylabel("Frequency")
plt.show()

# حساب متوسط الارتفاع لكل جنس
median_height_per_gender = df.groupby('latitude')['housing_median_age'].median()

# تعبئة القيم المفقودة في الارتفاع بالمتوسط لكل جنس
df['housing_median_age'] = df.apply(lambda row: median_height_per_gender[row['latitude']]
if pd.isnull(row['housing_median_age']) else row['housing_median_age'], axis=1)

# التحقق من عدم وجود قيم مفقودة في الارتفاع
print(f"Missing housing_median_age values after imputation: {df['housing_median_age'].isnull().sum()}")

# الخطوة 2: اكتشاف والتعامل مع القيم المتطرفة -----------------------------------------

# دالة لحساب حدود IQR لاكتشاف القيم المتطرفة
def detect_outliers_iqr(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

# الميزات التي نريد التحقق من القيم المتطرفة فيها
features = ['total_rooms', 'total_bedrooms', 'population']

# حساب الحدود لكل ميزة
bounds = {feature: detect_outliers_iqr(df, feature) for feature in features}

# تعديل الحدود الدنيا لضمان الاتساق المنطقي
bounds['total_rooms'] = (max(bounds['total_rooms'][0], 10), bounds['total_rooms'][1])  # يجب أن يكون عدد الغرف ≥ 10
bounds['total_bedrooms'] = (max(bounds['total_bedrooms'][0], 5), bounds['total_bedrooms'][1])  # يجب أن يكون عدد الغرف ≥ 5
bounds['population'] = (max(bounds['population'][0], 1), bounds['population'][1])  # يجب أن يكون عدد السكان ≥ 1

# عرض حدود القيم المتطرفة
for feature, (low, high) in bounds.items():
    print(f"{feature} outliers below {low:.2f} or above {high:.2f}")

# الخطوة 3: التعامل مع القيم المتطرفة ----------------------------------------------------

# دالة لإزالة أو استبدال القيم المتطرفة
def handle_outliers(df, feature, lower_bound, upper_bound, threshold=0.008):
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    outlier_percentage = len(outliers) / len(df)

    if outlier_percentage < threshold:
        print(f"Removing {len(outliers)} outliers from {feature} (below {threshold * 100}%)")
        df = df[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]
    else:
        print(f"Replacing outliers in {feature} with median")
        median_value = df[feature].median()
        df[feature] = np.where((df[feature] < lower_bound) | (df[feature] > upper_bound),
                               median_value, df[feature])

    return df

# تطبيق التعامل مع القيم المتطرفة لكل ميزة
for feature, (low, high) in bounds.items():
    df = handle_outliers(df, feature, low, high)

# الخطوة 4: التحقق النهائي ----------------------------------------------------

# التحقق من عدم وجود قيم مفقودة بعد المعالجة
print("\nMissing values after handling:\n", df.isnull().sum())

# عرض ملخص البيانات بعد التنظيف
print("\nDataset summary after cleaning:\n", df.describe())

# حفظ البيانات النظيفة
df.to_csv("result_of_california_housing_train.csv", index=False)
print("\nCleaned dataset saved successfully.")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

#  تحميل البيانات وتصحيح أسماء الأعمدة
df = pd.read_csv("/california_housing_train.csv")
df.columns = df.columns.str.lower().str.strip()  # جعل الأسماء بحروف صغيرة وإزالة المسافات

#  تحديد الميزات (features) والهدف (target)
features = df.drop(columns=['longitude', 'housing_median_age'])  # إزالة بعض الأعمدة
target = df['longitude']  # نستخدم longitude كمتغير مستهدف (Regression)

#  تقسيم البيانات إلى تدريب واختبار
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

#  إنشاء نموذج انحدار (Regression)
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)  # تدريب النموذج

#  توقع القيم
y_pred = model.predict(X_test)

#  تقييم النموذج باستخدام MSE و R²
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f" Baseline Model MSE: {mse:.4f}")
print(f" Baseline Model R² Score: {r2:.4f}")

#  دالة لاختيار الميزات بناءً على التباين
def apply_variance_threshold(threshold):
    selector = VarianceThreshold(threshold)
    selected_features = selector.fit_transform(features)

    retained_feature_names = features.columns[selector.get_support()]
    print(f"\n Threshold: {threshold}")
    print(f" Number of Features Before: {features.shape[1]}")
    print(f" Number of Features After: {len(retained_feature_names)}")
    print(f" Eliminated Features: {set(features.columns) - set(retained_feature_names)}")

    return selected_features, retained_feature_names

#  تجربة إزالة الميزات ذات التباين المنخفض واختبار تأثيرها على النموذج
for threshold in [0.0, 0.01, 0.05, 0.1]:
    selected_features, feature_names = apply_variance_threshold(threshold)

    # تقسيم البيانات بعد اختيار الميزات
    X_train_fs, X_test_fs, y_train, y_test = train_test_split(selected_features, target, test_size=0.2, random_state=42)

    # إنشاء وتدريب النموذج بعد تقليل الميزات
    model_fs = RandomForestRegressor(random_state=42)
    model_fs.fit(X_train_fs, y_train)

    # التوقع وتقييم الأداء
    y_pred_fs = model_fs.predict(X_test_fs)
    mse_fs = mean_squared_error(y_test, y_pred_fs)
    r2_fs = r2_score(y_test, y_pred_fs)

    print(f" Model MSE After Feature Selection (Threshold {threshold}): {mse_fs:.4f}")
    print(f" Model R² Score After Feature Selection (Threshold {threshold}): {r2_fs:.4f}")
```

تحديد عدد الميزات التي سيتم اختيارها (k) بناءً على أعلى قيم Information Gain.
في هذه التعديلات، سنقوم باستخدام SelectKBest مع mutual_info_classif بدلاً من VarianceThreshold لاختيار الميزات بناءً على مقدار المعلومات (Information Gain) التي توفرها كل ميزة بالنسبة للهدف (target).



```python
#2.5.2 Filtering Method using Information Gain
#Task 2.12: Modify the code snippet in the 'Filtering Method using Variance' section to
# demonstrate feature filtering using Information Gain. To calculate information gain,
# utilize the SelectKBest() method with the mutual_info_classif function in sklearn.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# د تحميل البيانات وتصحيح أسماء الأعمدة
df = pd.read_csv("/california_housing_train.csv")
df.columns = df.columns.str.lower().str.strip()  # جعل الأسماء بحروف صغيرة وإزالة المسافات

#  تحديد الميزات (features) والمتغير المستهدف (target)
features = df.drop(columns=['longitude', 'housing_median_age'])
target = (df['longitude'] > df['longitude'].median()).astype(int)  # تحويل `longitude` إلى فئتين (لتحويل المشكلة إلى تصنيف)

#  تقسيم البيانات إلى تدريب واختبار
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

#  إنشاء نموذج التصنيف الأساسي
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)  # تدريب النموذج

#  توقع القيم وتقييم الأداء
y_pred = model.predict(X_test)
baseline_accuracy = accuracy_score(y_test, y_pred)

print(f" Baseline Model Accuracy: {baseline_accuracy:.4f}")

#  دالة لاختيار الميزات بناءً على Information Gain
def apply_information_gain(k):
    selector = SelectKBest(score_func=mutual_info_classif, k=k)
    selected_features = selector.fit_transform(features, target)

    retained_feature_names = features.columns[selector.get_support()]
    print(f"\n Selecting Top {k} Features using Information Gain")
    print(f" Number of Features Before: {features.shape[1]}")
    print(f" Number of Features After: {len(retained_feature_names)}")
    print(f" Selected Features: {list(retained_feature_names)}")

    return selected_features, retained_feature_names

# تجربة اختيار الميزات باستخدام Information Gain
for k in [2, 4, 6]:  # اختيار أفضل 2، 4، و6 ميزات
    selected_features, feature_names = apply_information_gain(k)

    # تقسيم البيانات بعد اختيار الميزات
    X_train_fs, X_test_fs, y_train, y_test = train_test_split(selected_features, target, test_size=0.2, random_state=42)

    # إنشاء وتدريب النموذج بعد تقليل الميزات
    model_fs = RandomForestClassifier(random_state=42)
    model_fs.fit(X_train_fs, y_train)

    # التوقع وتقييم الأداء
    y_pred_fs = model_fs.predict(X_test_fs)
    accuracy_fs = accuracy_score(y_test, y_pred_fs)

    print(f" Model Accuracy After Feature Selection (Top {k} Features): {accuracy_fs:.4f}")
```
